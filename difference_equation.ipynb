```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y

    for i in range(len_y):
        for k in range(len_x):
            if(i - k >= 0 and i - k < len_h):
                y[i] += x[k] * h[i - k]

    return y
```

```python
x = np.array([1, 2, 3, 4, 5])
h = np.array([0.3, 0.4, 0.3])
y = np.convolve(h, x)
print('y(n) =', y)
```

```python
# ============================================
# Difference Equation Implementation
# ============================================
# A difference equation relates output y[n] to input x[n] and past values
# General form: y[n] = b0*x[n] + b1*x[n-1] + ... - a1*y[n-1] - a2*y[n-2] - ...

# The given system: y[n] = 0.3*x[n] + 0.4*x[n-1] + 0.3*x[n-2]
# This is a FIR (Finite Impulse Response) filter - Moving Average filter
# Coefficients: h = [0.3, 0.4, 0.3] represent the impulse response

# ============================================
# Input Signal
# ============================================
n_x = np.arange(len(x))  # Sample indices for input signal x[n]

# ============================================
# Output Signal (from convolution)
# ============================================
n_y = np.arange(len(y))  # Sample indices for output signal y[n]

# ============================================
# Plotting Input and Output Signals
# ============================================
fig, axes = plt.subplots(3, 1, figsize=(10, 8))

# Plot 1: Input Signal x[n]
axes[0].stem(n_x, x, basefmt='k-', linefmt='b-', markerfmt='bo')
axes[0].set_title('Input Signal x[n]', fontsize=12, fontweight='bold')
axes[0].set_xlabel('n (sample index)')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[0].set_xticks(n_x)

# Plot 2: Impulse Response h[n]
n_h = np.arange(len(h))  # Sample indices for impulse response
axes[1].stem(n_h, h, basefmt='k-', linefmt='g-', markerfmt='go')
axes[1].set_title('Impulse Response h[n] = [0.3, 0.4, 0.3]', fontsize=12, fontweight='bold')
axes[1].set_xlabel('n (sample index)')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True, linestyle='--', alpha=0.7)
axes[1].set_xticks(n_h)

# Plot 3: Output Signal y[n] = x[n] * h[n] (convolution)
axes[2].stem(n_y, y, basefmt='k-', linefmt='r-', markerfmt='ro')
axes[2].set_title('Output Signal y[n] = x[n] * h[n] (Convolution)', fontsize=12, fontweight='bold')
axes[2].set_xlabel('n (sample index)')
axes[2].set_ylabel('Amplitude')
axes[2].grid(True, linestyle='--', alpha=0.7)
axes[2].set_xticks(n_y)

plt.tight_layout()
plt.show()

# ============================================
# Display Results
# ============================================
print("=" * 50)
print("DIFFERENCE EQUATION ANALYSIS")
print("=" * 50)
print(f"\nInput signal x[n]: {x}")
print(f"Impulse response h[n]: {h}")
print(f"Output signal y[n]: {y}")
print(f"\nLength of x[n]: {len(x)}")
print(f"Length of h[n]: {len(h)}")
print(f"Length of y[n]: {len(y)} (= {len(x)} + {len(h)} - 1)")
print("\n" + "=" * 50)
print("The output y[n] is computed using convolution:")
print("y[n] =  x[k] * h[n-k]")
print("=" * 50)
```
