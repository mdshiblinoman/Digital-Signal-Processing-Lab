```python
# Import required libraries
# numpy: For numerical operations and array manipulation
# matplotlib.pyplot: For plotting and visualization
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sine Wave
# Generate a continuous sine wave signal
# t: time array from 0 to 1 second with 0.01 second steps
# y: sine wave with amplitude 2 and frequency 2 Hz
t = np.arange(0, 1, 0.01)
y = 2 * np.sin(2 * np.pi * 2 * t)

# Plot the sine wave
plt.plot(t, y)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python
# Cosine Wave
# Generate a continuous cosine wave signal
# t: time array from 0 to 1 second with 0.01 second steps
# y: cosine wave with amplitude 2 and frequency 2 Hz
t = np.arange(0, 1, 0.01)
y = 2 * np.cos(2 * np.pi * 2 * t)

# Plot the cosine wave
plt.plot(t, y)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.show()
```

```python
# Unit Impulse Sequence (Delta Function)
# (n) = 1 when n = shift, 0 otherwise
def unit_impulse(shift, n):
    """
    Generate a unit impulse (delta) sequence
    Parameters:
        shift: position where the impulse occurs
        n: array of sample indices
    Returns:
        delta: list containing the impulse sequence
    """
    delta = []
    for i in n:
        if(i == shift):
            delta.append(1)  # Impulse at shift position
        else:
            delta.append(0)  # Zero elsewhere
    return delta

# Main Function
# Create sample indices from -5 to 5
n = np.arange(-5, 6, 1)
# Generate unit impulse at n=0
delta = unit_impulse(0, n)

# Plot the unit impulse sequence using stem plot
plt.stem(n, delta)
plt.xlabel('n')
plt.ylabel('(n)')
plt.title('Unit Impulse Sequence')
plt.show()
```

```python
# Unit Step Sequence
# u(n) = 0 when n < shift, 1 when n >= shift
def unit_step(shift, n):
    """
    Generate a unit step sequence
    Parameters:
        shift: position where the step begins
        n: array of sample indices
    Returns:
        step: list containing the step sequence
    """
    step = []
    for i in n:
        if(i < shift):
            step.append(0)  # Zero before shift
        else:
            step.append(1)  # One at and after shift
    return step

# Main Function
# Create sample indices from -5 to 5
n = np.arange(-5, 6, 1)
# Generate unit step starting at n=0
step = unit_step(0, n)

# Plot the unit step sequence using stem plot
plt.stem(n, step)
plt.xlabel('(n)')
plt.ylabel('u(n)')
plt.title('Unit Step Sequence')
plt.show()
```

```python
# Unit Ramp Sequence
# ur(n) = 0 when n < shift, n when n >= shift
def unit_ramp(shift, n):
    """
    Generate a unit ramp sequence
    Parameters:
        shift: position where the ramp begins
        n: array of sample indices
    Returns:
        ramp: list containing the ramp sequence
    """
    ramp = []
    j = 0  # Counter for ramp values
    for i in n:
        if(i < shift):
            ramp.append(0)  # Zero before shift
        else:
            ramp.append(j)  # Linear increment at and after shift
            j += 1
    return ramp

# Main function
# Create sample indices from -5 to 5
n = np.arange(-5, 6, 1)
# Generate unit ramp starting at n=0
ramp = unit_ramp(0, n)

# Plot the unit ramp sequence using stem plot
plt.stem(n, ramp)
plt.xlabel('n')
plt.ylabel('ur(n)')
plt.title('Unit Ramp Signal')
plt.show()
```

```python
# Exponential Signal, when 0<a<1
# This represents a decaying exponential signal
# exp(n) = a^n where 0 < a < 1
n = np.arange(0, 20, 1)  # Sample indices from 0 to 19
a = 0.8  # Base value (decay factor)
exp = a ** n  # Calculate exponential values

# Plot the decaying exponential signal using stem plot
plt.stem(n, exp)
plt.xlabel('n')
plt.ylabel('exp(n)')
plt.title('Exponential Signal (0<a<1)')
plt.show()
```

```python
# Exponential Signal, when a>1
# This represents a growing exponential signal
# exp(n) = a^n where a > 1
n = np.arange(0, 20, 1)  # Sample indices from 0 to 19
a = 1.2  # Base value (growth factor)
exp = a ** n  # Calculate exponential values

# Plot the growing exponential signal using stem plot
plt.stem(n, exp)
plt.xlabel('n')
plt.ylabel('exp(n)')
plt.title('Exponential Signal (a>1)')
plt.show()
```
